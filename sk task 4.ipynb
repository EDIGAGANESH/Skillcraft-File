```python
print("\nColumn Names:\n", data.columns.tolist())

# Verify key columns
key_columns = ['Start_Time', 'Weather_Condition', 'Start_Lat', 'Start_Lng', 'Severity']
missing_columns = [col for col in key_columns if col not in data.columns]
if missing_columns:
    print(f"\nError: Columns {missing_columns} not found. Please check dataset columns and update code.")
    exit()
print("\nMissing Values in Key Columns:\n", data[key_columns].isnull().sum())
print("\nUnique Weather Conditions (Top 10):\n", data['Weather_Condition'].value_counts().head(10))
```

```python
#Preprocessing
data['Start_Time'] = pd.to_datetime(data['Start_Time'], errors='coerce')
data['Hour'] = data['Start_Time'].dt.hour
data = data.dropna(subset=key_columns)
relevant_columns = ['Start_Time', 'Hour', 'Weather_Condition', 'Start_Lat', 'Start_Lng', 'Severity', 'Traffic_Signal', 'Junction', 'Crossing']
available_columns = [col for col in relevant_columns if col in data.columns]
data = data[available_columns]
print("\nPreprocessed Dataset Shape:", data.shape)
print("\nPreprocessed Columns:\n", data.columns.tolist())
data.to_csv('preprocessed_accidents.csv', index=False)
print("\nPreprocessed data saved to 'preprocessed_accidents.csv'.")
```

```python
#Analysis
# Time of Day
hourly_accidents = data['Hour'].value_counts().sort_index()
print("\nHourly Accident Counts:\n", hourly_accidents)
```

```python
# Weather
weather_accidents = data['Weather_Condition'].value_counts().head(10)
print("\nTop 10 Weather Conditions:\n", weather_accidents)
```

```python
# Road Conditions (use Traffic_Signal if available, else Weather_Condition)
road_condition_column = 'Weather_Condition'
if 'Traffic_Signal' in data.columns:
    road_condition_column = 'Traffic_Signal'
    road_condition_accidents = data['Traffic_Signal'].value_counts()
else:
    road_condition_accidents = data['Weather_Condition'].value_counts().head(10)
```

```python
# Severity by Weather
severity_weather = data.groupby('Weather_Condition')['Severity'].mean().sort_values(ascending=False).head(10)
print("\nTop 10 Weather Conditions by Severity:\n", severity_weather)
```

```python
#Visualization
# Accidents by Hour
plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_accidents.index, y=hourly_accidents.values, color='skyblue')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.grid(axis='y')
plt.show()
```

```python
# Accidents by Weather Condition
plt.figure(figsize=(12, 6))
sns.barplot(x=weather_accidents.values, y=weather_accidents.index, color='lightgreen')
plt.title('Top 10 Weather Conditions for Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Condition')
plt.grid(axis='x')
plt.show()
```

```python
# Accidents by Road Condition
plt.figure(figsize=(12, 6))
sns.barplot(x=road_condition_accidents.values, y=road_condition_accidents.index.astype(str), color='lightblue')
plt.title(f'Accidents by Road Condition ({road_condition_column} as Proxy)')
plt.xlabel('Number of Accidents')
plt.ylabel('Condition')
plt.grid(axis='x')
plt.show()
```

```python
# Accident Hotspots (Scatter Plot)
plt.figure(figsize=(12, 8))
sample_data = data.sample(10000, random_state=42)
plt.scatter(sample_data['Start_Lng'], sample_data['Start_Lat'], s=2, alpha=0.2, c='red')
plt.title('Accident Hotspots (Start_Lat vs Start_Lng)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# Severity by Weather
plt.figure(figsize=(12, 6))
sns.barplot(x=severity_weather.values, y=severity_weather.index, color='salmon')
plt.title('Average Accident Severity by Weather Condition')
plt.xlabel('Average Severity (1-4)')
plt.ylabel('Weather Condition')
plt.grid(axis='x')
plt.show()
```
